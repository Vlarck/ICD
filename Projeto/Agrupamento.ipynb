```python
import pandas as pd
df = pd.read_csv("Life Expectancy Data.csv")
df
```

```python
nulos_por_media = df.fillna(df.mean())
nulos_por_media
```

```python
import seaborn as sns
sns.scatterplot(data=nulos_por_media, x="Income composition of resources", y="Schooling")
```

```python
from sklearn.cluster import KMeans
```

```python
df_group = nulos_por_media[["Income composition of resources", "Schooling"]]
```

### Definindo o valor de K

```python
kmeans = KMeans(n_clusters=2)
```

### Gerando os grupos 

```python
kmeans.fit(df_group)
```

### Demonstração dos grupos a partir das posições

```python
kmeans.labels_
```

### Gráfico de agrupamento

```python
sns.scatterplot(data=nulos_por_media, y="Income composition of resources", x="Schooling", hue=kmeans.labels_)
```

### Países com pessoas que passam mais tempo na escola, tendem a possuírem um maior índice de desenvolvimento humano (IDH)

### Analisando outras duas variáveis ( IMC e expectativa de vida )

```python
import seaborn as sns
sns.scatterplot(data=nulos_por_media, x=" BMI ", y="Life expectancy ")
```

```python
df_group2 = nulos_por_media[[" BMI ", "Life expectancy "]]
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(df_group2)
```

```python
sns.scatterplot(data=nulos_por_media, x=" BMI ", y="Life expectancy ", hue=kmeans.labels_)
```

### Nesse caso, nota-se uma proporção direta entre o IMC e a expctativa de vida. O mais lógico não seria uma proporção inversa entre essas variáveis?

### Analisando expectativa de vida e mortalidade adulta

```python
import seaborn as sns
sns.scatterplot(data=nulos_por_media, x="Life expectancy ", y="Adult Mortality")
```

```python
df_group3 = nulos_por_media[["Adult Mortality", "Life expectancy "]]
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df_group3)
```

```python
sns.scatterplot(data=nulos_por_media, x="Life expectancy ", y="Adult Mortality", hue=kmeans.labels_)
```

### Neste outro caso, nota-se uma relação lógica entre a expectativa de vida e a mortalidade adulta, sendo variáveis inversamente proporcionais.
